```python
import os
import numpy as np
import pandas as pd
from datetime import date
from datetime import datetime
```

```python
# Load 2 csv files
df1 = pd.read_csv("data_2023-02-11.csv", sep= ';')
df2 = pd.read_csv("data_2023-02-12.csv", sep= ';')
```

```python
# Read the dataset 1
df1.head()
```

```python
# Read the dataset 2
df2.head()
```

```python
# Convert "metric_date" column to datetime
df1['metric_date'] = pd.to_datetime(df1['metric_date'])
df2['metric_date'] = pd.to_datetime(df2['metric_date'])
```

```python
# Combine the data frames
combined_data = pd.concat([df1,df2])
```

```python
# Read the combined dataset
combined_data
```

```python
# Add a column for the combined file generation date
current_datetime = datetime.now()
# Compact datatime of combined_file_generation_date with DATE only
formatted_datetime = datetime(current_datetime.year, current_datetime.month, current_datetime.day, current_datetime.hour, current_datetime.minute)
# Assign the formatted_datetime to additional columnn in combined_data
combined_data['combined_file_generation_date'] = formatted_datetime
```

```python
# Read the combined data
combined_data
```

```python
# Output the combined data to a single CSV file
combined_data.to_csv('combined_weekend_data', index = False)
print("Combined data saved successfully to 'combined_weekend_data.csv'")
```
